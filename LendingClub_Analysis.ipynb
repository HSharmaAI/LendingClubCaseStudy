```python
import pandas as pd

# Load the data
loan_data = pd.read_csv('loan.csv', low_memory=False)
data_dictionary = pd.read_excel('Data_Dictionary.xlsx')

# Remove columns with all missing values
loan_data.dropna(axis=1, how='all', inplace=True)

# Separate numeric and non-numeric columns
numeric_columns = loan_data.select_dtypes(include=['number']).columns
non_numeric_columns = loan_data.select_dtypes(exclude=['number']).columns

# Fill missing values for numeric columns with median
loan_data[numeric_columns] = loan_data[numeric_columns].apply(lambda x: x.fillna(x.median()))

# Fill missing values for non-numeric columns with mode (most frequent value)
for column in non_numeric_columns:
    loan_data[column] = loan_data[column].fillna(loan_data[column].mode()[0])

# Verify no missing values remain
missing_values = loan_data.isnull().sum()
print(missing_values[missing_values > 0])

# Handling outliers
Q1 = loan_data[numeric_columns].quantile(0.25)
Q3 = loan_data[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Filter out the outliers for each numeric column
loan_data = loan_data[~((loan_data[numeric_columns] < (Q1 - 1.5 * IQR)) | (loan_data[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Verify data shape after removing outliers
print(loan_data.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of loan amount
plt.figure(figsize=(10, 6))
sns.histplot(loan_data['loan_amnt'], bins=30, kde=True)
plt.title('Distribution of Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()

# Analyze and visualize categorical features
plt.figure(figsize=(10, 6))
sns.countplot(data=loan_data, x='grade', order=loan_data['grade'].value_counts().index)
plt.title('Distribution of Loan Grades')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the relationship between loan amount and loan status
plt.figure(figsize=(10, 6))
sns.boxplot(data=loan_data, x='loan_status', y='loan_amnt')
plt.title('Loan Amount by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Loan Amount')
plt.show()

# Analyze the effect of annual income on loan status
plt.figure(figsize=(10, 6))
sns.boxplot(data=loan_data, x='loan_status', y='annual_inc')
plt.title('Annual Income by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Annual Income')
plt.show()
```

```python
# Correlation matrix for numeric columns
plt.figure(figsize=(12, 8))
correlation_matrix = loan_data[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
